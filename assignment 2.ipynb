```python
# prompt: Using any in-built library present in Python, count the occurrences of the words in the text (.csv) and give the â€˜Top 30â€™ most common words.

from collections import Counter

# Read the text from the CSV file
with open('/content/CSV2.csv', 'r') as f:
  text = f.read()

# Split the text into words
words = text.split()

# Count the occurrences of each word
word_counts = Counter(words)

# Get the top 30 most common words
top_30_words = word_counts.most_common(30)

# Print the top 30 most common words
for word, count in top_30_words:
  print(f"{word}: {count}")
```

Using the â€˜Auto Tokenizerâ€™ function in the â€˜Transformersâ€™ library, write a â€˜functionâ€™ to count unique tokens in the text (.csv) and give the â€˜Top 30â€™ words.

```python
# prompt: Using the â€˜Auto Tokenizerâ€™ function in the â€˜Transformersâ€™ library, write a â€˜functionâ€™ to count unique tokens in the text (.csv) and give the â€˜Top 30â€™ words.

from transformers import AutoTokenizer

def count_unique_tokens(text):
  """
  Counts the unique tokens in a text using the 'Auto Tokenizer' function in the 'Transformers' library.

  Args:
    text: The text to be analyzed.

  Returns:
    A list of the top 30 most common tokens and their counts.
  """

  tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
  tokens = tokenizer.tokenize(text)
  unique_tokens = set(tokens)
  token_counts = Counter(tokens)
  top_30_tokens = token_counts.most_common(30)
  return top_30_tokens

# Read the text from the CSV file
with open('/content/CSV2.csv', 'r') as f:
  text = f.read()

# Get the top 30 most common tokens
top_30_tokens = count_unique_tokens(text)

# Print the top 30 most common tokens
for token, count in top_30_tokens:
  print(f"{token}: {count}")

```

```python
# prompt: Extract the â€˜diseasesâ€™ entities in the â€˜.csv fileâ€™

import pandas as pd

# Load the CSV file into a Pandas DataFrame
df = pd.read_csv('/content/CSV2.csv')

# Extract the 'diseases' entities using spaCy
diseases = []
for text in df['TEXT']:
  doc = nlp(text)
  for entity in doc.ents:
    if entity.label_ == 'DISEASE':
      diseases.append(entity.text)

# Print the extracted diseases
for disease in diseases:
  print(disease)
```
